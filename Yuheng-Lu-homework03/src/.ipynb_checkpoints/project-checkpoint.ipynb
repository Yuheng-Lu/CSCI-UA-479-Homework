#### Type Conversion

```python
DataFrame = DataFrame[["started_at", "ended_at", "start_station_name", "end_station_name"]]
DataFrame["started_at"] = pd.to_datetime(DataFrame["started_at"], format="%Y-%m-%d %H:%M:%S")
DataFrame["ended_at"] = pd.to_datetime(DataFrame["ended_at"], format="%Y-%m-%d %H:%M:%S")
```

1. I get the columns I need out.
2. I convert two columns which store date time as string to datetime64 type.

#### Create New Calculated Columns

```python
import numpy as np
from scipy import stats
DataFrame["duration"] = DataFrame["ended_at"] - DataFrame["started_at"]
print("Raw Summary of Riding Duration:")
print(DataFrame["duration"].describe(), "\n")
DataFrame["duration_int"] = (DataFrame["duration"] / np.timedelta64(1, 's')).astype(int)
'''
I did not use .dt.seconds attribute in pandas to convert timedelta to integer,
because I found .dt.seconds has upper and lower bounds (0s to 86399s(23h59min59s))
'''
DataFrame_trimmed = DataFrame[DataFrame["duration_int"] >= 0]
print("Summary of Riding Duration")
print(DataFrame_trimmed["duration"].describe())

z_scores_duration = stats.zscore(DataFrame_trimmed["duration_int"])
abs_z_scores_duration = np.abs(z_scores_duration)
duration_int_trimmed = DataFrame_trimmed[abs_z_scores_duration<3]["duration_int"]
```

1. Use describe() to get a summary of the raw data.
2. There are several negative values of duration, which is not possible; otherwise, they will be time traveling events, so I get rid of them. However, it is hard to make sure whether data like 4 days are valid or not. These data look like errors but not impossible. I do not think I should get rid of them arbitrarily.
3. Finally, I use z-score to trim down data considered outliers.

#### Visualization 1

```python
import matplotlib.pyplot as plt
plt.boxplot(duration_int_trimmed, sym = '+', vert = False)
plt.title("Biking Duration Box Plot")
plt.xlabel("Biking Duration(s)")
plt.show()

plt.violinplot(duration_int_trimmed, vert = False)
plt.title("Biking Duration Violin Plot")
plt.xlabel("Biking Duration(s)")
plt.show()
```

1. Boxplot of biking duration.
2. Violin plot of biking duration, which is more beautiful, but lack of information like mean, 25 percent quantile, 75 percent quantile, etc.

#### Calculate Value Counts


```python
start_station_name_value_counts = DataFrame["start_station_name"].value_counts()
end_station_name_value_counts = DataFrame["end_station_name"].value_counts()
```

#### Visualization 2

```python
plt.bar(start_station_name_value_counts.index[:10], start_station_name_value_counts[:10])
plt.xticks(rotation='vertical')
plt.title("Top 10 Station Names VS Numbers Out (Start)")
plt.xlabel("Station Names")
plt.ylabel("Numbers Out (Start)")
plt.show()

plt.bar(end_station_name_value_counts.index[:10], end_station_name_value_counts[:10])
plt.xticks(rotation='vertical')
plt.title("Top 10 Station Names VS Numbers In (End)")
plt.xlabel("Station Names")
plt.ylabel("Numbers In (End)")
plt.show()
```

1. Value counts of the times of existence of different station names.
