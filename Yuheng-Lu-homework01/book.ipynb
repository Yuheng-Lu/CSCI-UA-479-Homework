### My Question:
- What is the most commonly used pronoun(s) in *Alice's Adventures in Wonderland* and *Pride and Prejudice*?

```python
word_count_dict = {}

with open("data/Alice‚Äôs Adventures in Wonderland.txt", "r") as book1:
    for line in book1:
        lst_line = line.split(" ")
        for word in lst_line:
            clean_word = ""
            for char in word:
                if char.isalpha():
                    clean_word += char
            word = clean_word.lower()           
            if word in word_count_dict.keys() and word != "":
                word_count_dict[word] += 1
            elif word != "":
                word_count_dict[word] = 1

with open("data/Pride and Prejudice.txt", "r") as book2:
    for line in book2:
        lst_line = line.split(" ")
        for word in lst_line:
            clean_word = ""
            for char in word:
                if char.isalpha():
                    clean_word += char
            word = clean_word.lower()           
            if word in word_count_dict.keys() and word != "":
                word_count_dict[word] += 1
            elif word != "":
                word_count_dict[word] = 1 

word_count_dict = {key: value for key, value in sorted(word_count_dict.items(), key=lambda item: item[1])}
print(word_count_dict)
```

### Steps:
- First, I created an empty dictionary counts all words in books. The keys are words, and the values are numbers.
- Second, I opened book #1, which is *Pride and Prejudice*. For each lines in the book, I splited the line into a list of words and clean the words because there might include some "\n" or symbols we do not want. For each cleaned words, if the word has already stored in the dictionary, I would increment the value by one; if not, I would create the pair.
- Third, I opened book #2, which is *Alice's Adventures in Wonderland*. For each lines in the book, I splited the line into a list of words and clean the words because there might include some "\n" or symbols we do not want. For each cleaned words, if the word has already stored in the dictionary, I would increment the value of it by one; if not, I would create the pair.
- Fourth, I sorted the dictionary by values.
- It shows that the most commonly used pronoun is "her", which appears 2464 times!
